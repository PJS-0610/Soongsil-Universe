```
import random
import time
import sys
def main():
  global count
  global display
  global word
  global already_guessed
  global length
  global play_game
  global a
  words_to_guess=["mouse","izone","friend","omygirl","phone","computer","eraser","pencil","favorite","airpods","backpack","soldier","midterm","curtain","masterpiece","iloveyou"]
  word=random.choice(words_to_guess)
  a=word
  length=len(word)
  count=0
  display="_"*length
  already_guessed=[]

def play_loop():
  global play_game
  play_game=input("게임을 다시 하고 싶으신가요? y = yes, n = no\n")
  while play_game not in ["y", "Y", "n", "N"]:
    play_game=input("게임을 다시 하고 싶으신가요? y = yes, n = no\n")
  if play_game=='y' or play_game=='Y':
    main()
    hangman()
    quit()
  elif play_game=='n' or play_game=='N':
    print("게임을 즐겨주셔서 감사합니다! 다음에 또 뵙기를 바라겠습니다!\n")
    sys.exit()

def hangman():
  global count
  global display
  global word
  global already_guessed
  global play_game
  limit=5
  while True:
    guess=input('\n게임 제시 단어입니다: '+display+" 당신의 예상을 입력하세요: \n")
    guess=guess.strip()
    if len(guess.strip())==0 or len(guess.strip())>=2:
      print("잘못된 입력입니다. 다시 입력해주세요\n")
      hangman()
    elif guess in word:
      already_guessed.extend(guess)
      index=word.find(guess)
      word=word[:index]+"_"+word[index+1:]
      display=display[:index]+guess+display[index+1:]
      print(display+"\n")
    elif guess in already_guessed:
      print('이미 하신 글자입니다.\n')
    else:
      count=count+1
      if count==1:
        time.sleep(1)
        print("\n    ______ \n"
              "   |      \n"
              "   |      \n"
              "   |      \n"
              "   |      \n"
              "   |      \n"
              "   |      \n"
              "___|___\n")
        print("\n틀렸습니다. 남은 횟수는",limit - count,"번")
      elif count==2:
        time.sleep(1)
        print("    ______ \n"
              "   |      | \n"
              "   |      |\n"
              "   |      \n"
              "   |      \n"
              "   |      \n"
              "   |      \n"
              "___|___\n")
        print("\n틀렸습니다. 남은 횟수는",limit - count,"번")
      elif count==3:
        time.sleep(1)
        print("    ______ \n"
              "   |      | \n"
              "   |      |\n"
              "   |      |\n"
              "   |      \n"
              "   |      \n"
              "   |      \n"
              "___|___\n")
        print("\n틀렸습니다. 남은 횟수는",limit - count,"번")
      elif count==4:
        time.sleep(1)
        print("    ______ \n"
              "   |      | \n"
              "   |      |\n"
              "   |      |\n"
              "   |      O\n"
              "   |      \n"
              "   |      \n"
              "___|___\n")
        print("\n틀렸습니다. 남은 횟수는",limit - count,"번")
      elif count==5:
        time.sleep(1)
        print("    ______ \n"
              "   |      | \n"
              "   |      |\n"
              "   |      |\n"
              "   |      O\n"
              "   |     -|-\n"
              "   |      |\n"
              "___|___\n")
        print("틀렸습니다. 사망....")
        print("정답은",a,"입니다\n")
        play_loop()
        quit()

    if len(word)==len(already_guessed):
      print("축하드립니다 정답입니다.",display,"\n")
      play_loop()
      quit()
  
  




print('행맨 게임에 오신 것을 환영합니다!')
name=input("이름을 입력하세요: ")
print("안녕하세요",name,"님 행운을 빕니다!")
time.sleep(1)
print("게임이 곧 시작됩니다!\n행맨 게임을 즐기세요!")
time.sleep(2)
main()
hangman()
```
